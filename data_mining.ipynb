```python
# prompt: import pandas untuk pd

import pandas as pd
```

```python
df = pd.read_excel('/content/IP___Riwayat_Medical_01012016 s.d 19032024 - Copy.xls')
df.head()
```

```python
# prompt: Ubah semua data menjadi lowercase (casefolding) termasuk header

df.columns = df.columns.str.lower()
for column in df.columns:
  if pd.api.types.is_string_dtype(df[column]):
    df[column] = df[column].str.lower()

df.head()
```

```python
# prompt: Revisi nama fitur menjadi lebih umum, casefolding, jika terdiri dari 2 kata, sambung
# dengan underscored ( _ )

def revisi_nama_fitur(nama_fitur):
  """Revisi nama fitur menjadi lebih umum, casefolding, jika terdiri dari 2 kata, sambung dengan underscored."""
  nama_fitur = nama_fitur.lower()
  nama_fitur = nama_fitur.replace(" ", "_")
  return nama_fitur


new_columns = [revisi_nama_fitur(col) for col in df.columns]
df.columns = new_columns

df.head()
```

```python
import matplotlib.pyplot as plt

for column in df.columns:
  plt.figure(figsize=(15, 6))  # Atur lebar dan panjang diagram
  plt.hist(df[column], bins=20)  # Buat histogram dengan 20 bins
  plt.title(f"Histogram of {column}")
  plt.xlabel(column)
  plt.ylabel("Frequency")
  plt.show()
```

```python
for column in df.columns:
  plt.figure(figsize=(10, 6))
  plt.scatter(df.index, df[column])
  plt.title(f"Scatter Plot of {column}")
  plt.xlabel("Index")
  plt.ylabel(column)
  plt.show()
```
